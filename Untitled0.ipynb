```python
from clarifai.client import ClarifaiApi
import nltk, random
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir('.') if isfile(join('.', f))]
api = ClarifaiApi();
imgs = []
for f in onlyfiles:
    if 'jpg' in f:
        imgs.append('./' + f)
```

```python
results = []

for i in imgs[6:8]:
    results.append(api.tag_images([open(i)]))
```

```python
results
```

```python
words = results[1]['results'][0]['result']['tag']['classes']
words += results[0]['results'][0]['result']['tag']['classes']
```

```python
words
```

```python
sentence = ''

for w in words:
    sentence += str(w) + ' '
```

```python
sentence
```

```python
text = nltk.word_tokenize(sentence)
```

```python
tags = nltk.pos_tag(text)
```

```python
tags
```

```python
nouns = []
adjectives = []
plural = []
verbs = []

for t in tags:
    if 'NNS' in t[1]:
        plural.append(t[0])
    elif 'JJ' in t[1]:
        adjectives.append(t[0])
    elif 'NN' in t[1]:
        nouns.append(t[0])
    elif 'V' in t[1]:
        verbs.append(t[0])
```
